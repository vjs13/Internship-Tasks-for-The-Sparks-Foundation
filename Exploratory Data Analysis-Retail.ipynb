```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
print('Libraries imported')
```

```python
data=pd.read_csv('C:\\Users\\Vikram\\Downloads\\SampleSuperstore.csv')
data.head()
```

```python
data.info
```

```python
#Removing duplicate values:
data=data.drop_duplicates()
```

```python
#Analyzing the data
print('Dimensions of Dataset: {}'.format(data.shape),'\n','-' * 75)
print('Number of Duplicated values: {}'.format(data.duplicated().sum()),'\n','-' * 75)
print('Null values contained by rows: \n{}'.format(data.isnull().sum()),'\n','-'*75)
print('Schema of dataset: \n',data.info(),'\n','-'*75)
print('Descriptive analysis of data: \n{}'.format(data.describe(include='all')),'\n','-'*75)
print('Describing correlation of data: \n{}'.format(data.corr()),'\n','-'*75)
```

# EXPLORATORY DATA ANALYSIS

```python
#correlation matrix
plt.figure(figsize=(8,5))
correlation=data.corr()
sns.heatmap(correlation,annot=True,cmap='Blues')
```

```python
plt.figure()
sns.pairplot(data)
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='Discount',y='Profit',data=data)
plt.legend('Profit')
plt.grid(alpha=0.3)
```

```python
plt.figure(figsize=(30,13))
plt.plot(data['Profit'],color='Blue')
plt.grid(alpha=0.7)
```

```python
#Distribution of Segments
%matplotlib inline
sns.countplot('Segment',data=data)
plt.xlabel('SEGMENT',fontsize=14)
plt.ylabel('COUNT',fontsize=14)
plt.title('DISTRIBUTION OF SEGMENTS')
plt.grid(alpha=0.2)
```

## From the above graph we can conclude that the most goods are bought by "Consumers" and least are bought by "Home Office"

```python
#Distribution of Ship Mode
%matplotlib inline
sns.countplot('Ship Mode',data=data)
plt.xlabel('SHIP MODE',fontsize=12)
plt.ylabel('COUNT',fontsize=12)
plt.title('DISTRIBUTION OF SHIP MODE')
```

## From the above graph we can conclude that the most number of deliveries are done in 'Standard Class' and least number of deliveries are done 'Same Day'.

```python
#Distribution of Shipments based on States
%matplotlib inline
plt.figure(figsize=(50,10))
sns.countplot('State',data=data)
plt.xticks(fontsize=20,rotation=90)
plt.xlabel('States',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.title('Distribution of Shipments based on States',fontsize=12)
```

## From the graph above we can see that most customers are from 'California' and least customers are from 'West Virginia'

```python
#Distribution based on region 
%matplotlib inline
sns.countplot(data['Region'])
plt.xlabel('REGION',fontsize=13)
plt.ylabel('COUNT',fontsize=13)
plt.title('DISTRIBUTION BASED ON REGION',fontsize=15)
```

## From the above graph we can observe that the most orders are comming from 'WEST' of United States

```python
#Distribution based on Category
%matplotlib inline
sns.countplot(data['Category'])
plt.xlabel('CATEGORY',fontsize=14)
plt.ylabel('COUNT',fontsize=14)
plt.title('DISTRIBUTION BASED ON CATEGORY',fontsize=15)
plt.show()
```

## From the above graph we can observe that the most number of goods which are bought are from the 'Office Supplies' category.

```python
#Distribution of Sub Categories
%matplotlib inline
plt.figure(figsize=(30,10))
plt.xticks(fontsize=15)
sns.countplot(data['Sub-Category'])
plt.xlabel('Sub-Category',fontsize=17)
plt.ylabel('COUNT',fontsize=17)
plt.title('DISTRIBUTION BASED ON Sub-Category',fontsize=16)
plt.show()
```

## From here we can observe that the product which is sold the most is 'Binders' and the one which is sold the least is 'Copiers'

```python
#Distribution of Sub-categories among categories
plt.figure(figsize=(18,10))
sns.countplot('Category',hue='Sub-Category',data=data);
plt.title('DISTRIBUTION OF SUB-CATEGORIES AMONG CATEGORIES')
plt.xlabel('SUB-CATEGORY')
plt.ylabel('COUNT')
plt.grid(alpha=0.3)
```

## From the graph above we can observe that in 'Furniture', 'Furnishings' are sold the most.
## In 'Office-Supplies', 'Binders' are sold the most.
## In 'Technology', 'Phones' are sold the most.

```python
#Profit distribution amongst states
%matplotlib inline
plt.figure(figsize=(30,10))
sns.barplot(x=data['State'].sort_values(ascending=True),y=data['Profit'])
plt.xticks(fontsize=14,rotation=90)
plt.xlabel('STATE',fontsize=19)
plt.ylabel('PROFIT',fontsize=19)
plt.title('STATES VS PROFIT',fontsize=19)
```

```python
%matplotlib inline
C=data[data['Region']=='Central']
plt.figure(figsize=(30,10))
sns.barplot(x=C['State'].sort_values(ascending=True),y=C['Profit'])
plt.xticks(fontsize=15,rotation=90)
```
